```python
#Importar librerias necesarias
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# 1. Importar data frames
df_company = pd.read_csv('moved_project_sql_result_01.csv')
df_location = pd.read_csv('moved_project_sql_result_04.csv')
# 2. Analizar data frames
df_company.info()
print()
df_location.info() 
```

```python
# 3. Transformar el average_trip  int
df_location['average_trips'] = df_location['average_trips'].round().astype(int)
```

```python
# 4. Organizar y graficar el data frame basado en average_trip los top 10 

df_order_location = df_location.sort_values(by='average_trips', ascending=False).head(10)

pivot_df = df_order_location.pivot( #Cambiar la forma de la tabla para graficar
    #index='average_trips',
    columns='dropoff_location_name',
    values='average_trips'
)

pivot_df.plot(kind='bar', figsize=(14, 6), rot=0) #rot=0 para mantener los numeros de los meses horizontal
plt.xlabel('Dropoff Location name')
plt.ylabel('Average trips quantity')
plt.title('Average trips per Location')
plt.legend(title='Location')
plt.tight_layout()
plt.show()

```
